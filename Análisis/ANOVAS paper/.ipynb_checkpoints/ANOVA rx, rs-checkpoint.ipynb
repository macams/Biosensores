```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import colors
from scipy.interpolate import UnivariateSpline
from scipy.optimize import curve_fit
from scipy import stats
import seaborn as sns
```

```python
rrx=[54.98,17.82,27.2]
yrx=[0.39,75.68,23.93]
rrs=[58.75,24.31,16.94]
yrs=[2.09,88.40,9.51]
```

```python
factores=['RFP TU','YFP TU','Other']
colores=['red','yellow','lightgray']
factac = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(factores, rrx)]

plt.figure(figsize=(6,6))
plt.title('RFP rx',loc=('left'),fontsize=30)
patches, texts = plt.pie(rrx, colors=colores, pctdistance=0.6,labeldistance=1.1,radius=1.,startangle=90)
plt.axis('equal')
plt.legend(patches, factac, loc=(1.01,0.7))
plt.savefig(fname='anova RFPrx.png',dpi=1200,facecolor='w',edgecolor='w',bbox_inches='tight',pad_inches=0.2)
```

```python
factac = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(factores, yrx)]

plt.figure(figsize=(6,6))
plt.title('YFP rx',loc=('left'),fontsize=30)
patches, texts = plt.pie(yrx, colors=colores, pctdistance=0.6,labeldistance=1.1,radius=1.,startangle=90)
plt.axis('equal')
plt.legend(patches, factac, loc=(1.01,0.7))
plt.savefig(fname='anova YFPrx.png',dpi=1200,facecolor='w',edgecolor='w',bbox_inches='tight',pad_inches=0.2)
```

```python
factac = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(factores, rrs)]

plt.figure(figsize=(6,6))
plt.title('RFP rs',loc=('left'),fontsize=30)
patches, texts = plt.pie(rrs, colors=colores, pctdistance=0.6,labeldistance=1.1,radius=1.,startangle=90)
plt.axis('equal')
plt.legend(patches, factac, loc=(1.01,0.7))
plt.savefig(fname='anova RFPrs.png',dpi=1200,facecolor='w',edgecolor='w',bbox_inches='tight',pad_inches=0.2)
```

```python

factac = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(factores, yrs)]

plt.figure(figsize=(6,6))
plt.title('YFP rs',loc=('left'),fontsize=30)
patches, texts = plt.pie(yrs, colors=colores, pctdistance=0.6,labeldistance=1.1,radius=1.,startangle=90)
plt.axis('equal')
plt.legend(patches, factac, loc=(1.01,0.7))
plt.savefig(fname='anova YFPrs.png',dpi=1200,facecolor='w',edgecolor='w',bbox_inches='tight',pad_inches=0.2)
```

```python

yname = ['YFP rs','YFP rx','RFP rs','RFP rx' ]
r=np.array([2.09,0.39,58.75,54.98])
y=np.array([88.40,75.68,24.31,17.82])
o=np.array([9.51,23.93,16.94,27.2])


plt.title('Analysis of variance (ANOVA)',fontsize=15)
plt.xlabel('Percentage (%)')
plt.barh(yname, r, color='red',label='RFP TU')
plt.barh(yname, y, left=r, color='yellow',label='YFP TU')
plt.barh(yname,o, left=r+y, color='lightgray',label='Other')

plt.legend(loc=(1.01,0.0))

plt.savefig(fname='anovabarplot.png',dpi=1200,facecolor='w',edgecolor='w',bbox_inches='tight',pad_inches=0.2)
```
