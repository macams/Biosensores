```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import pandas.io.sql as psql
import numpy as np
import scipy
from scipy.optimize import curve_fit
from matplotlib import colors
import seaborn as sns
```

```python
vrfp=[27.77,2.26,50.60,19.37]
factores=['Plásmido','Medio','Inductor','Otro']
colores=['royalblue','dimgray','teal','limegreen']
factac = ['{0} - {1:1.2f} %'.format(i,j) for i,j in zip(factores, vrfp)]

plt.figure(figsize=(6,6))
plt.title('RFP',loc=('left'),fontsize=30)
patches, texts = plt.pie(vrfp, colors=colores, pctdistance=0.6,labeldistance=1.1,radius=1.,startangle=90)
plt.axis('equal')
plt.legend(patches,factac,loc=(1.01,0.7))
#plt.savefig(fname='anova rfp entropia.png',dpi=1200,facecolor='w',edgecolor='w',bbox_inches='tight',pad_inches=0.2)
```
